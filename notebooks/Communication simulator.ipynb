```python
class Node:

    outgoing_byte_rate: int
    incoming_byte_rate: int
    outgoing_byte_count: int
    incoming_byte_count: int
    
    __slots__ = [
        'outgoing_byte_rate',
        'incoming_byte_rate',
        'outgoing_byte_count',
        'incoming_byte_count'
    ]

    def init(self):
        self.outgoing_byte_rate = 3 * (10 ** 8) # 300MB/s
        self.incoming_byte_rate = 3 * (10 ** 8) # 300MB/s
        self.outgoing_byte_count = 0
        self.incoming_byte_count = 0


class Analyst(Node):
    
    schedulers: [Scheduler]
    
    __slots__ = [
        'schedulers',
        'outgoing_byte_rate',
        'incoming_byte_rate',
        'outgoing_byte_count',
        'incoming_byte_count'
    ]


class Scheduler(Node):
    
    parent_analyst: Analyst
    executors: [Executor]
    
    __slots__ = [
        'parent_analyst',
        'executors',
        'outgoing_byte_rate',
        'incoming_byte_rate',
        'outgoing_byte_count',
        'incoming_byte_count'
    ]


class Executor(Node):
    
    parent_scheduler: Scheduler
    FLOPS: int
    
    __slots__ = [
        'parent_scheduler',
        'FLOPS',
        'outgoing_byte_rate',
        'incoming_byte_rate',
        'outgoing_byte_count',
        'incoming_byte_count'
    ]


model_layer_dims = [784, 1024, 2048, 1024, 512, 10]
model_layer_count = len(model_layer_dims)
# input_feature_size = model_layer_dims[0]
# output_class_size = model_layer_[-1]
epochs = 1000
options = {'use_global_FLOPS': False}


# class Simulator:
#
#     __slots__ = [
#         'analyst_node',
#         'scheduler_nodes'
#     ]
```
