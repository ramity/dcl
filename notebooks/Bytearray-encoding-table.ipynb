# Checking how numpy encodes arrays to bytes

```python
import base64
import time

occurences = [0 for _ in range(256)]
size = [1024, 1024]
random_array = numpy.random.rand(*size)
random_array_bytes = random_array.tobytes()
random_array_bytes_len = len(random_array_bytes)

start = time.perf_counter_ns()
encoded_bytes = base64.b64encode(random_array_bytes)
end = time.perf_counter_ns()
delta = end - start
delta_seconds = delta / (10 ** 9)
rate = random_array_bytes_len / delta
rate_seconds = random_array_bytes_len / delta_seconds

print('delta ns: {}'.format(delta))
print('delta s: {}'.format(delta_seconds))
print('rate bytes/ns {}'.format(rate))
print('rate bytes/s {}'.format(rate_seconds))
print('rate kb/s {}'.format((random_array_bytes_len / (10 ** 3)) / delta_seconds))
print('rate mb/s {}'.format((random_array_bytes_len / (10 ** 6)) / delta_seconds))
print('rate gb/s {}'.format((random_array_bytes_len / (10 ** 9)) / delta_seconds))

for array_byte in encoded_bytes:
    occurences[array_byte] += 1
print(occurences)
```

# Simulation of the buffer_updated callback

```python
import secrets
import re


# Create a convincing receive buffer with byte sequence patterns

receive_buffer_size = 10 ** 9
receive_buffer_bytes = bytearray(secrets.token_bytes(receive_buffer_size))
receive_buffer = memoryview(receive_buffer_bytes)
receive_buffer[0:9] = bytearray("<message>", "utf8")
receive_buffer[10000:10010] = bytearray("</message>", "utf8")
receive_buffer[10010:10019] = bytearray("<message>", "utf8")
receive_buffer[20000:20010] = bytearray("</message>", "utf8")
receive_buffer[20010:20019] = bytearray("<message>", "utf8")
receive_buffer[30000:30010] = bytearray("</message>", "utf8")
receive_buffer[30010:30019] = bytearray("<message>", "utf8")
receive_buffer[40000:40010] = bytearray("</message>", "utf8")
receive_buffer[40010:40019] = bytearray("<message>", "utf8")
receive_buffer[-10:] = bytearray("</message>", "utf8")


# These variables are expected to be on the protocol object
# - last_message_offset denotes the position after the last paired close tag

last_message_offset = 0
open_message_byte_string = b"<message>"
open_message_byte_size = len(open_message_byte_string)
close_message_byte_string = b"</message>"
close_message_byte_size = len(close_message_byte_string)
buffer_offset = 0


# These variables are calculated/provided every buffer_updated call to determine what part of the receieve_buffer is considered

nbytes = 30025
search_space_start = last_message_offset
search_space_end = buffer_offset + nbytes


# Look through a porition of the recieve buffer (simulate buffer_update callback)
# - In this case, the first 30k bytes have been recieved
#   - Simulates three open and closed messages with a fourth unclosed message
# - In other cases, the start is always the index after the last processed close message tag

pending_open_message_tag_positions = [_.start() for _ in re.finditer(open_message_byte_string, receive_buffer[search_space_start:search_space_end])]
pending_close_message_tag_positions = [_.start() for _ in re.finditer(close_message_byte_string, receive_buffer[search_space_start:search_space_end])]
pending_open_message_tag_count = len(pending_open_message_tag_positions)
pending_close_message_tag_count = len(pending_close_message_tag_positions)
pending_messages = pending_close_message_tag_count

# Detect if there isn't any work to be completed
# - Handles case that the space updated before the buffer_updated call is only composed of message data

if pending_messages == 0:
    
    # Don't forget to add the nbytes to the buffer_offset for next buffer_updated call
    buffer_offset += nbytes
    pass
    # return True


# Handle processing buffer L to R
# - Key to note that there will always more opens than closes
# - Also to note that the beginning will always be a open message tag
# * This loop uses a local variable to save the last applicible_end_message_position

for i in range(pending_close_message_tag_count):
    
    applicible_start_message_position = pending_open_message_tag_positions[i]
    applicible_end_message_position = pending_close_message_tag_positions[i] + close_message_byte_size
    
    print("The following message is pushed to the task queue:")
    print("offset {}:{}".format(applicible_start_message_position, applicible_end_message_position))
    print(receive_buffer_bytes[applicible_start_message_position:applicible_end_message_position])
    print("-" * 100)


# Lastly, update the buffer_offset and set the last_message_offset to the last processed close tag

buffer_offset += nbytes
last_message_offset = pending_close_message_tag_positions[-1]
```

# Check fastest way to detect if a substring exists in a buffer

```python
import time

start: int = 0
end: int = 0
delta: int = 0
count: int = 100000
delta_sum: int = 0
```

```python
for i in range(count):
    start = time.perf_counter_ns()
    if bytearray('</message>', 'utf8') in receive_buffer_bytes[search_space_start:search_space_end]:
        end = time.perf_counter_ns()
        delta = end - start
        delta_sum += delta

print('AVG {} seconds'.format((delta_sum / count) / 10 ** 9))
delta_sum = 0
```

```python
for i in range(count):
    start = time.perf_counter_ns()
    if bytes('</message>', 'utf8') in receive_buffer_bytes[search_space_start:search_space_end]:
        end = time.perf_counter_ns()
        delta = end - start
        delta_sum += delta

print('AVG {} seconds'.format((delta_sum / count) / 10 ** 9))
delta_sum = 0
```

```python
open_message_byte_pattern = re.compile(open_message_byte_string)

for i in range(count):
    start = time.perf_counter_ns()
    if re.search(open_message_byte_pattern, receive_buffer[search_space_start:search_space_end]):
        end = time.perf_counter_ns()
        delta = end - start
        delta_sum += delta

print('AVG {} seconds'.format((delta_sum / count) / 10 ** 9))
delta_sum = 0
```

```python
for i in range(count):
    start = time.perf_counter_ns()
    if re.match(open_message_byte_pattern, receive_buffer[search_space_start:search_space_end]):
        end = time.perf_counter_ns()
        delta = end - start
        delta_sum += delta

print('AVG {} seconds'.format((delta_sum / count) / 10 ** 9))
delta_sum = 0
```

```python
for i in range(count):
    start = time.perf_counter_ns()
    if re.finditer(open_message_byte_pattern, receive_buffer[search_space_start:search_space_end]):
        end = time.perf_counter_ns()
        delta = end - start
        delta_sum += delta

print('AVG {} seconds'.format((delta_sum / count) / 10 ** 9))
delta_sum = 0
```

```python
for i in range(count):
    start = time.perf_counter_ns()
    if re.findall(open_message_byte_pattern, receive_buffer[search_space_start:search_space_end]):
        end = time.perf_counter_ns()
        delta = end - start
        delta_sum += delta

print('AVG {} seconds'.format((delta_sum / count) / 10 ** 9))
delta_sum = 0
```

```python
for i in range(count):
    start = time.perf_counter_ns()
    if re.split(open_message_byte_pattern, receive_buffer[search_space_start:search_space_end]):
        end = time.perf_counter_ns()
        delta = end - start
        delta_sum += delta

print('AVG {} seconds'.format((delta_sum / count) / 10 ** 9))
delta_sum = 0
```
