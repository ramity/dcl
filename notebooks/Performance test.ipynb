```python
import numpy
import time
import csv

tests = 100

start_base = 2
end_base = 14

sizes = [2 ** x for x in range(start_base, end_base + 1)]

with open('metrics.csv', 'w', newline='') as csv_file:
    
    csv_writer = csv.writer(csv_file, delimiter=',')
    field_names = ['a element count', 'b element count', 'a & b memory (bytes)', 'add FLOPS', 'sub FLOPS', 'mul FLOPS', 'div FLOPS', 'dot FLOPS']
    csv_writer.writerow(field_names)
    
    for size in sizes:

        a = numpy.random.rand(size)
        b = numpy.random.rand(size)

        time_sum = 0
        for test in range(tests):
            start_time = time.perf_counter_ns()
            c = a + b
            end_time = time.perf_counter_ns()
            delta_time = end_time - start_time
            time_sum += delta_time
        average_time = time_sum / tests
        operations = size
        add_flops = operations / (average_time / (10 ** 9))

        time_sum = 0
        for test in range(tests):
            start_time = time.perf_counter_ns()
            c = a - b
            end_time = time.perf_counter_ns()
            delta_time = end_time - start_time
            time_sum += delta_time
        average_time = time_sum / tests
        operations = size
        sub_flops = operations / (average_time / (10 ** 9))

        time_sum = 0
        for test in range(tests):
            start_time = time.perf_counter_ns()
            c = a * b
            end_time = time.perf_counter_ns()
            delta_time = end_time - start_time
            time_sum += delta_time
        average_time = time_sum / tests
        operations = size
        mul_flops = operations / (average_time / (10 ** 9))

        time_sum = 0
        for test in range(tests):
            start_time = time.perf_counter_ns()
            c = a / b
            end_time = time.perf_counter_ns()
            delta_time = end_time - start_time
            time_sum += delta_time
        average_time = time_sum / tests
        operations = size
        div_flops = operations / (average_time / (10 ** 9))

        time_sum = 0
        a_dot = a.reshape(size, 1)
        b_dot = b.reshape(1, size)
        for test in range(tests):
            start_time = time.perf_counter_ns()
            c = numpy.dot(a_dot, b_dot)
            end_time = time.perf_counter_ns()
            delta_time = end_time - start_time
            time_sum += delta_time
        average_time = time_sum / tests
        operations = size * size
        dot_flops = operations / (average_time / (10 ** 9))

        byte_count = (size * 8) * 2
        csv_writer.writerow([size, size, byte_count, add_flops, sub_flops, mul_flops, div_flops, dot_flops])
        csv_file.flush()
        
        del a
        del b
        del a_dot
        del b_dot
```
